```python
# Importing dependencies

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sqlalchemy import create_engine
import numpy as np

# Creating engine to work with

engine = create_engine(f'postgresql://test:test@localhost:5432/SQL Homework')
connection = engine.connect()
```

```python
# Importing dataframes from SQL for analysis

# employee df
employees = pd.read_sql('SELECT * FROM employees', connection)
# salaries df
salaries = pd.read_sql('SELECT * FROM salaries', connection)
# titles df
titles = pd.read_sql('SELECT * FROM titles', connection)
```

```python
# Printing employees df

employees.head(5)
```

```python
# Printing salaries df

salaries.head(5)
```

```python
# Printing salaries df

titles.head(5)
```

```python
# Renaming column names for titles df for future merge

titles_clean=titles.rename(columns={"title_id":"emp_title_id"})

# Printing df

titles_clean.head(5)
```

```python
# Merging both of the dataframes: employees & salaries df's

emp_salary=pd.merge(employees, salaries, on="emp_no", how="left")

# Printing df

emp_salary.head(5)
```

```python
# Merging both of the dataframes: emp_salary & titles_clean df's

emp_salary_title=pd.merge(emp_salary,titles_clean, on="emp_title_id")

# Printing amount number of employee

print(f'There are {len(emp_salary_title)} employees in the company.')

# Printing df

emp_salary_title.head(5)
```

```python
# Create a histogram to visualize the most common salary ranges for employees

emp_salary_title.hist(column='salary', bins=20, figsize=(5,5), color='#86bf91', rwidth=0.9)
plt.xlabel('Salary Ranges')
plt.ylabel('Frecuency of Salaries')
plt.title('Frecuency of Salaries Plot')

# Printing plot

plt.show()
```

```python
# Grouping by Titles to get average salaries with inplace method

title_summary=emp_salary_title.groupby(['title']).mean() # avgerage salaries

title_summary.drop('emp_no', axis=1, inplace=True) # dropping columns

title_summary=title_summary.rename(columns={"salary":"Salary"}) # renaming columns

# Printing df

title_summary.head(5)
```

```python
# Bar Chart of average salary by title

title_summary=emp_salary_title["title"].value_counts()
plt.bar(title_summary.index.values, title_summary.values)
title_summary.plot(kind="bar")
plt.xlabel("Job Title")
plt.xticks(rotation=90)
plt.ylabel("Salary in USD")
plt.title("Average Salary Per Job Title")

plt.show()
```
